```python
y = df2['Humidity']
x = df2.drop(['Formatted Date','Summary','Humidity','Daily Summary'], axis=1)
x
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_x_0 = LabelEncoder()
x['Precip Type'] = labelencoder_x_0.fit_transform(x['Precip Type'])
x
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
predictor = Sequential()
predictor.add(Dense(6, activation='relu', kernel_initializer='uniform',input_dim=8))
predictor.add(Dense(6, activation = 'relu', kernel_initializer = 'uniform'))
predictor.add(Dense(1, activation = 'sigmoid', kernel_initializer = 'uniform'))
```

```python
predictor.compile(optimizer = 'adam', loss = 'mean_squared_error',  metrics=['MeanSquaredError'])
```

```python
from sklearn.metrics import mean_squared_error
meanSqauredErrors = []

for i in range(10):
  predictor.fit(X_train, y_train, batch_size = 10)
  y_predi = predictor.predict(X_test)
  y_pre = pd.DataFrame(y_predi)
  meanSqauredErrors.append(mean_squared_error(y_pre, y_test))
meanSqauredErrors

```

```python
import matplotlib.pyplot as plt
plt.plot(meanSqauredErrors)
plt.xlabel('Epoch number')
plt.ylabel('Mean Squared Error')
```
